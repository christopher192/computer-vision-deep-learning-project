## OCR with Pytesseract

```python
# import libraries
import cv2
import pytesseract
import matplotlib.pyplot as plt
```

```python
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

#### 1. Simple Usage

```python
image = cv2.imread("image/address.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
text = pytesseract.image_to_string(image)
print(text)
```

#### 2. Detecting and OCR`s Digits

```python
image = cv2.imread("image/apple_support.png")
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
# extract and print a string of digits from an image
# the outputbase digits option tells tesseract to treat the image as a single line of text containing only digits
options = "outputbase digits"
text = pytesseract.image_to_string(rgb, config = options)

print(text)
```

#### 3. Whitelisting and Blacklisting Characters for OCR

```python
image = cv2.imread("image/pa_license_plate.png")
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
# blacklist the *# character
options = "-c tessedit_char_blacklist=*#"
text = pytesseract.image_to_string(rgb, config = options)

print(text)
```

```python
image = cv2.imread("image/invoice.png")
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
# whitelist the 0123456789.- character
options = "-c tessedit_char_whitelist=0123456789.-"
text = pytesseract.image_to_string(rgb, config = options)

print(text)
```
